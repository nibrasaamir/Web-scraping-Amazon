```python
import pandas as pd
import json

jfile = 'data.json'
with open(jfile, 'r') as file:
    jdata = json.load(file)
df = pd.json_normalize(jdata)
df
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={
    'productId': 'ASIN',
    'product reviews': 'Reviews',
    'productUrl': 'URL',
    'productName': 'title',
    'images.src': 'images',
    'colors and images.src': 'color_images',
    'colors and images.alt': 'colors'
})
df
```

```python
columns_to_clean = ['ASIN', 'gender', 'manufacturer']

for column in columns_to_clean:
    df[column] = df[column].apply(lambda x: x.split(': ', 1)[-1])
    
df    
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df
```

```python
df['review_texts'] = df['Reviews'].apply(lambda reviews: [review['review'] for review in reviews])
df
```

```python
print(df['review_texts'].loc[0])
```

```python
import pandas as pd
import re

def clean_text(text):
    if not isinstance(text, str):
        text = str(text)
    
    # Removing URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Keeping only words, spaces, commas, and periods
    cleaned_text = re.sub(r'[^\w\s,\.]', '', text)
    # Removing commas and periods
    cleaned_text = cleaned_text.replace(',', '')
    cleaned_text = cleaned_text.replace('.', '')
    # Convert to lowercase
    return cleaned_text.lower()

df['review_texts'] = df['review_texts'].apply(clean_text)
df['description'] = df['description'].apply(clean_text)
df
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
def remove_stop_words(text):
    tokens = word_tokenize(text)
    filtered_tokens = [token for token in tokens if token.lower() not in
    stop_words]
    return ' '.join(filtered_tokens)

df['review_texts'] = df['review_texts'].apply(remove_stop_words)
df['description'] = df['description'].apply(remove_stop_words)
df
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token, pos='v') for token in tokens]
    return ' '.join(lemmatized_tokens)

df['review_texts'] = df['review_texts'].apply(lemmatize_text)
df['description'] = df['description'].apply(lemmatize_text)
df
```

```python
df = df.drop('Reviews', axis=1)
df
```

```python
import pandas as pd
import os
import requests
from PIL import Image
from io import BytesIO


def download_image(url, save_path):
    try:
        response = requests.get(url, timeout=10)  # 10 seconds timeout
        if response.status_code == 200:
            image = Image.open(BytesIO(response.content))
            image.save(save_path)
            print(f"Downloaded and saved: {save_path}")
        else:
            print(f"Failed to download {url}: HTTP {response.status_code}")
    except Exception as e:
        print(f"Error downloading {url}: {e}")

def process_dataframe(df):
    for _, row in df.iterrows():
        product_id = row['ASIN']
        for column_name in ['images', 'color_images']:
            image_list = row[column_name]
            folder_path = os.path.join('downloaded_images', product_id, column_name)
            os.makedirs(folder_path, exist_ok=True)
            
            for i, url in enumerate(image_list):
                file_name = f"{i}.jpg"
                save_path = os.path.join(folder_path, file_name)
                download_image(url, save_path)

process_dataframe(df)
```

```python
# products(asin,  gender, rating, price, decription, brandname, color)
# amazon(url, mnufacturer, title)
# category(category, subcategory)
# images(color images, images)
# reviews(text)
df = df.drop_duplicates(subset=['ASIN'], keep='first')
```

```python
df['amazon_id'] = ['A' + str(i).zfill(1) for i in range(len(df))]
df['cat_id'] = ['X' + str(i).zfill(1) for i in range(len(df))]
df['image_id'] = ['N' + str(i).zfill(1) for i in range(len(df))]
df['rev_id'] = ['M' + str(i).zfill(1) for i in range(len(df))]
```

```python
product = df[['ASIN','gender','rating','price','description','brandName','colors','cat_id','image_id','rev_id']]
amazon = df[['amazon_id','URL','manufacturer','title','ASIN']]
category = df[['cat_id','category','subCategory']]
images = df[['image_id','color_images','images']]
reviews = df[['rev_id','review_texts']]
```

```python
import pandas as pd
import pymysql

# Connect to the MySQL database
connection = pymysql.connect(host='localhost', user='root', password='12345678', database='amazon')
cursor = connection.cursor()

for i, r in product.iterrows():
    qu = "INSERT INTO product (ASIN, gender, rating, price, description, brandName, colors, cat_id, image_id, rev_id) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"  
    val = (str(r['ASIN']), str(r['gender']), float(r['rating']), float(r['price']), str(r['description']), str(r['brandName']), str(r['colors']), str(r['cat_id']), str(r['image_id']), str(r['rev_id'])) 
    cursor.execute(qu, val)
    
for i, r in amazon.iterrows():
    qu = "INSERT INTO amazon (amazon_id, URL, manufacturer, title, ASIN) VALUES (%s, %s, %s, %s, %s)"  
    val = (str(r['amazon_id']), str(r['URL']), str(r['manufacturer']), str(r['title']), str(r['ASIN'])) 
    cursor.execute(qu, val)
    
for i, r in category.iterrows():
    qu = "INSERT INTO category (cat_id, category, subCategory) VALUES (%s, %s, %s)"  
    val = (str(r['cat_id']), str(r['category']), str(r['subCategory'])) 
    cursor.execute(qu, val) 
    
for i, r in images.iterrows():
    qu = "INSERT INTO images (image_id, color_images, images) VALUES (%s, %s, %s)"  
    val = (str(r['image_id']), str(r['color_images']), str(r['images']))
    cursor.execute(qu, val)
    
for i, r in reviews.iterrows():
    qu = "INSERT INTO reviews (rev_id, review_texts) VALUES (%s, %s)"  
    val = (str(r['rev_id']), str(r['review_texts']))
    cursor.execute(qu, val)    

connection.commit()
connection.close()

```
