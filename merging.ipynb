```python
import pandas as pd

# Specify the path to your CSV file
file_path = 'df_mergedd.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Now df is a DataFrame containing the data from your CSV file.
# You can view the first few rows of the DataFrame using df.head()
df
```

```python
import pandas as pd
import json



# Initialize an empty list to hold the JSON object for each row
products_data = []

# Iterate over DataFrame rows as (index, Series) pairs
for i, row in df.iterrows():
    product_data = {
        "productUrl": row['productUrl'],
        "productId": row['productId'],
        "gender": row['gender'], 
        "manufacturer": row['manufacturer'],
        "category": row['category'],
        "description": row['description'],
        "subCategory": row['subCategory'],
        "price": row['price'],
        "productName": row['productName'],
        "brandName": row['brandName'],
        "rating": row['rating'],
        "images": row['images.src'],
        "colors and images": row['colors and images.src'],
        "colors": row['colors and images.alt'],
        "product reviews": row['product reviews']
    }
    products_data.append(product_data)

# Convert the list of product data to a JSON string
json_data = json.dumps(products_data, indent=4)

# Print or save your JSON data

with open('2.json', 'w') as f:
    f.write(json_data)
```
